```python
import chromadb
```

```python
chroma_client=chromadb.Client()
collection=chroma_client.get_or_create_collection(name="document")
```

```python


collection.add(
    ids=["id1", "id2","id3"],
    documents=[
        "This is a document about Pineapple",
        "This is a document about oranges",
        "I am a fruit that starts with O and end with s",
    ]
)
```

```python
from pprint import pprint
results = collection.query(
    query_texts=["This is fruit with yellow color"], # Chroma will embed this for you
    n_results=2, # how many results to return
    # where_document={'$contains': 'pineapple'} # document should contain this word
)
pprint(results)
```
